```python
import pandas as pd
```

```python
df= pd.read_csv('country_vaccinations_3.csv')
```

```python
df
```

```python
lst=[]
for i in df['country'].unique():
    lst.append([i, df[df['country']==i]['total_vaccinations'].sum()])
```

```python
df_= pd.DataFrame(lst, columns=['Country', 'total_vaccinations'])
```

```python
df_=df_.sort_values(by='total_vaccinations')
```

```python
least= df_['Country'].iloc[:1].values[0]
```

```python
print('The least number of total vaccinations are seen in : ', least)
```

```python
most=df_[-1:]['Country'].values[0]
```

```python
print('The most number of total vaccinations are seen in : ', most)
```
