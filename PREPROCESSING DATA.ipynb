```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('country_vaccinations_1.csv')
```

```python
df.head()
```

## ADDING DATE, MONTH AND YEAR COLUMNS

```python
year=[]
month=[]
date_=[]
for i in df['date']:
    year.append(i.split('-')[0])
    month.append(i.split('-')[1])
    date_.append(i.split('-')[2])
    
```

```python
len(year), len(month), len(date_)
```

```python
df['Year']= year
df['Month']= month
df['Date']=date_
```

```python
df
```

## ADDING DAY COLUMN

```python
import datetime
```

```python
day=[]
for i in df['date']:
    if(i.split('-')[1][0]!='0'):
        
        day.append(datetime.date(int(i.split('-')[0]),int(i.split('-')[1]),int(i.split('-')[2])).strftime('%a'))
    else:
        day.append(datetime.date(int(i.split('-')[0]),int(i.split('-')[1][1]), int(i.split('-')[2])).strftime('%a'))
    
    
    
```

```python
df['Day']=day
```

```python
df
```

```python
df.to_csv('country_vaccinations_2.csv', index= False)
```
