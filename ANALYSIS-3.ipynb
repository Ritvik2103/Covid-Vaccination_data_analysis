```python
import pandas as pd
```

```python
df= pd.read_csv('country_vaccinations_3.csv')
```

```python
df_=df[df['Year']==2021]
```

```python
df_
```

```python
lst=[]
for i in df_['month'].unique():
    lst.append([i, df_[df_['month']==i]['total_vaccinations'].sum()])
```

```python
df_1= pd.DataFrame(lst, columns=['Month', 'Vaccines'])
```

```python
df_1
```

```python
for i in df_1.values:
    print( 'Number of vaccines in the month:', i[0], 'are:' , i[1])
```
